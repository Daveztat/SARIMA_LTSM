```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_process import ArmaProcess

%matplotlib inline 
```

```python
np.random.seed(42) 
ma2 = np.array([1, 0.9, 0.3])
ar2 = np.array([1, 0, 0])

MA2_process = ArmaProcess(ar2, ma2).generate_sample(nsample=1000)
```

```python
fig, ax = plt.subplots()

ax.plot(MA2_process)
```
